```
import pandas as pd

news_train = pd.read_csv("tamil_news_train.csv", header=0, names=['article','headline','category','category_ta'])
news_train = news_train[['headline', 'category']]
news_train.category = pd.Categorical(news_train.category)
news_train['category'] = news_train['category'].cat.codes
news_train = news_train.dropna()
news_train.head()
```

```
news_test = pd.read_csv("./tamil_news_test.csv", header=0, names=['article','headline','category','category_ta'])
news_test = news_test[['headline', 'category']]
news_test.category = pd.Categorical(news_test.category)
news_test['category'] = news_test['category'].cat.codes
news_test = news_test.dropna()
news_test.head()
```

```
len(news_train.category.unique())
```

## News Categories

### TaMillion

```
from simpletransformers.classification import ClassificationModel
```

```
# set use_cuda=False on CPU-only platforms
model = ClassificationModel('bert', 'monsoon-nlp/tamillion', num_labels=6, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True
})
model.train_model(news_train)
```

Does it work?

```
result, model_outputs, wrong_predictions = model.eval_model(news_test)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("wrong predictions:")
print(str(len(wrong_predictions)) + ' wrong out of ' + str(len(news_test)))
bads
```

```
(3631-941)/3631*100
```

#### 74.1%

### Compare to mBERT

```
bert = ClassificationModel('bert', 'bert-base-multilingual-uncased', num_labels=6, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True
})
bert.train_model(news_train)
```

```
result, model_outputs, wrong_predictions = bert.eval_model(news_test)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("wrong predictions:")
print(str(len(wrong_predictions)) + ' wrong out of ' + str(len(news_test)))
bads
```

## Movie Reviews (lower error is better)
This has quarter and half star ratings, so I will consider it a regression task

```
! head -n 3 ./tamil_movie_reviews_train.csv
```

```
import pandas as pd 
movie_train = pd.read_csv("./tamil_movie_reviews_train.csv", header=0, names=['id','review','rating'])
movie_train = movie_train[['review', 'rating']]
movie_train.head()
```

```
movie_test = pd.read_csv("./tamil_movie_reviews_test.csv", header=0, names=['id','review','rating'])
movie_test = movie_test[['review', 'rating']]
movie_test.head()
```

### TaMillion

```
model = ClassificationModel('bert', 'monsoon-nlp/tamillion', num_labels=1, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True,
    'regression': True
})
model.train_model(movie_train)
result, model_outputs, wrong_predictions = model.eval_model(movie_test)
```

How to measure RMSE on regression tasks

```
from math import sqrt
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(movie_test.rating, model_outputs)
rmse = sqrt(mse)
rmse
```

#### 0.632
